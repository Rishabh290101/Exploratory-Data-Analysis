```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

print('Import successfull')
```

```python
from sklearn import datasets
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data,columns= ['sepal_length','sepal_width','petal_length','petal_width'])
iris_df['target'] = pd.Series(iris.target)
```

```python
pd.options.display.max_rows=500

iris_df.head(500)
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width',data = iris_df)
```

```python
sns.scatterplot(x='petal_length', y='petal_width',data = iris_df,hue='target')
```

```python
sns.set_style('whitegrid');
sns.FacetGrid(iris_df, hue='target', size=4).map(plt.scatter, 'sepal_length','sepal_width').add_legend();
plt.show();
```

```python
iris_df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width')
```

```python
new_df = iris_df.drop(['target'],axis=1,)columns=.iloc[0]
new_df.plot.pie(figsize= (5,5))
```

```python
iris_df['target'].value_counts().plot.pie()
```

```python
from pandas.plotting import andrews_curves
andrews_curves(iris_df,'sepal_width')
```

```python
iris_df.drop(['target'],axis=1).plot(subplots=True ,layout= (4,5), figsize=(16,8))
plt.tight_layout()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],color='black')
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='kde',fill=True)
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='reg')
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='hex')
```

```python
sns.pairplot(iris_df, hue="target");
plt.show()
```

```python
sns.boxplot(x='target',y='petal_length',data=iris_df)
plt.show()
```
